```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
plt.plot_date(dates, y, linestyle = 'solid')

#Using the to 'AutoDateFormatter' to automatically format dates on the x-axis: rotates the date to fit and change their alignments.
# plt.gcf() stands for "get current figure." It is a function that returns a reference to the current figure object.
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')
# Setting the " mpl_dates.DateFormatter" for the x-axis.
# plt.gca() stands for "get current axes." It is a function that returns a reference to the current Axes object.
plt.gca().xaxis.set_major_formatter(date_format) 
```

## More on plt.gcf() and plt.gca()

### plt.gcf()

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Plot some data on the axis
ax.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

# Get a reference to the current figure
current_fig = plt.gcf()

# Modify the current figure properties
current_fig.suptitle('My Plot')
current_fig.savefig('my_plot.png')

# Display the plot
plt.show()
```

### plt.gca()

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Get a reference to the current axes
current_ax = plt.gca()

# Plot some data on the axis
current_ax.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

# Customize the plot properties
current_ax.set_xlabel('X')
current_ax.set_ylabel('Y')
current_ax.set_title('My Plot')

# Display the plot
plt.show()
```
