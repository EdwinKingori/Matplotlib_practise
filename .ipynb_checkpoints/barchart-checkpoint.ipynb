## Objective: plotting popular languages according the the data submitted. 

### Using the csv module from the standard library to read the data.csv file and iterate through rows

### Using the built_in class Counters from the collections library to count the number of time each language has been used by the respondents. 



```python
with open ('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter ()
#getting/Counting data for all rows
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_counter)

# Printing the top most 15 used languages
print(language_counter.most_common(15))

#data for a single row
    #row = next(csv_reader)
    #print(row['LanguagesWorkedWith'].split(';'))
```

## PLoting the data from the turple given of the top 15 most used languages.
### Step 1 : Using a for loop to split the languages into their own list, and the counts into their own list in order to put them into the x and y axis.  
### Step 2: Plotting after assigning the x and y axis to languages and popularity


```python
#Step 1
languages = []
popularity = []

for i in language_counter.most_common (15):
    languages.append(i[0])
    popularity.append(i[1])
    
print(languages)
print(popularity)

```

```python
#Step 2: plotting
#using the reverse method to show the popular languages in desceding order

languages.reverse()
popularity.reverse()


plt.barh(languages, popularity) 

plt.title('The 15 Most Popular Langauages')
#plt.ylabel('Programming Langauages')
plt.xlabel('Number of users')

plt.tight_layout()
plt.grid(color= 'grey', linestyle = '-')

```
