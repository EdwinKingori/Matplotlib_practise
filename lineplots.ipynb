```python
data = pd.read_csv("data_2.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript'] 

plt.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#Using the fill_between
overall_median = 77616.0
plt.fill_between(ages, py_salaries, overall_median, 
                 where = (py_salaries > overall_median), interpolate = True,alpha = 0.25, label = "Above Average")

plt.fill_between(ages, py_salaries, overall_median, 
                 where = (py_salaries < overall_median), interpolate = True,alpha = 0.25, label = "Below Average")


plt.legend()

#plt.tight_layout()
plt.title("Median Salary (USD) by Age")
plt.xlabel('Ages')
plt.ylabel("Median Salary(USD)")

plt.style.use('fivethirtyeight')
plt.grid(linestyle = '--')
```
