```python
from matplotlib import pyplot as plt
```

### Creating a Piechart

```python
plt.style.use('fivethirtyeight')
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
#using an explode list to put an emphasis on a particular slice
explode = [0, 0 ,0, 0.1,0]
```

```python
#Adding arguments to enhance aesthtics, for instance autopct includes the percentages of each slice, 
plt.pie(slices,explode = explode, shadow = True, startangle =90,autopct = '%1.1f%%', labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('Language Popularity (Top 5)')
plt.tight_layout()
plt.show()
```
